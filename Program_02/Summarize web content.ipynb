Please use python to write a small program that allows users to enter any URL, comprehensively presents and summarizes the page content of this URL, and please deploy it online for testing.

```python
# importing necessery packages
import requests
from bs4 import BeautifulSoup
from gensim.summarization import summarize
```

```python
def get_page_content(url):
    try:
        # Fetch the HTML content of the URL
        response = requests.get(url)
        response.raise_for_status()  # Check for any request errors

        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract text content from HTML
        text_content = soup.get_text(separator='\n')

        return text_content
    except requests.exceptions.RequestException as e:
        return f"Error fetching content: {e}"
```

```python
def summarize_text(text):
    # Use Gensim's TextRank to generate a summary
    summary = summarize(text, ratio=0.3, word_count=None, split=False)

    return summary
```

```python
def main():
    # Get URL from the user
    url = input("Enter the URL: ")

    # Fetch and get the page content
    page_content = get_page_content(url)

    if page_content:
        # Display the page content
        print("\nPage Content:")
        print(page_content)

        # Summarize the page content
        summary = summarize_text(page_content)

        # Display the summary
        print("\nSummary:")
        print(summary)
    else:
        print("Failed to fetch page content.")

if __name__ == "__main__":
    main()
```
